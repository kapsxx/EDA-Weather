### Let's begin playing with the data!

- From the above data let me first consider the columns : "New Delhi Temperature [2 m elevation corrected]" and "New Delhi Relative Humidity [2 m]" i.e., "Temperature" and "Relative Humidity". Before finding a relationship between "Temperature" and "Relative Humidity" let us first study something about them.

- Absolute humidity is the mass of water vapor divided by the mass of dry air in a volume of air at a given temperature. Relative humidity is the ratio of the current absolute humidity to the highest possible absolute humidity.

```python
plt.figure(figsize = (30,10))

plt.title('Temperature and Relative Humidity', fontsize = 30)
plt.xlabel('Date', fontsize = 16)

plt.plot(df['timestamp'], df['New Delhi Temperature [2 m elevation corrected]'], color = 'orange', linewidth = 4, label = 'Temp')
plt.ylabel('Temperature (°C)', fontsize = 16)
plt.ylim([22, 90])

plt.legend(fontsize = 20, loc = 'upper left')

plt.twinx().plot(df['timestamp'], df['New Delhi Relative Humidity [2 m]'], color = 'cyan', linewidth = 4, label = 'Humidity')
plt.ylabel('Humidity (%)', fontsize = 16)

plt.legend(fontsize = 20, loc = (0.006,0.835))


plt.annotate('Max Temperature', fontstyle = 'italic', xy = (mdates.date2num(df.loc[15, 'timestamp']), 42.6973), xytext = (mdates.date2num(df.loc[1, 'timestamp']), 63),
             arrowprops = dict(facecolor = 'black', shrink = 0.035), fontsize = 20)
plt.annotate('Min Relative Humidity', fontstyle = 'italic', xy = (mdates.date2num(df.loc[15, 'timestamp']), 25), xytext = (mdates.date2num(df.loc[46, 'timestamp']), 72),
             arrowprops = dict(arrowstyle = '-|>', facecolor = 'black', connectionstyle = 'angle3, angleA=2, angleB=24'), fontsize = 20)

plt.grid(alpha = 0.3, color = 'grey')

plt.show()
```

- From the above plot we can clearly see that there is an inverse relationship between "Temperature" and "Relative Humidity". Also we see that 03-Aug-2020 witnessed the maximum tempaerature of 42.7 °C along with minimum relative humidity of 25% at 15:00:00 hrs.


- Moving on we can clearly see that there are various attributes that affect weather and also various factors that are closely related to each other. The next attribute is "New Delhi Precipitation Total" i.e., "Total Precipitation". Normally less than 2mm per hour is considered as "Slight Shower", whereas it is considered as "Moderate Shower" when the total precipitation amount is in the range betwenn 2mm to 10mm per hour and above 10mm, it is considered as "Heavy Shower".

```python
plt.figure(figsize = (30,10))

plt.bar(df['timestamp'], df['New Delhi Precipitation Total'], width = 0.028, label = 'Precipitation (mm)', color = 'fuchsia')

plt.title('Total Precipitation ', fontsize = 30)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Total Precipitation (mm)', fontsize = 15)

plt.ylim(0,11)

plt.axhspan(2, 10, color = 'cyan', alpha = 0.1, label = 'Area of Moderate Rainfall')
plt.axhline(y = 2, color = 'red', linestyle = ':', linewidth = 2)

plt.legend(loc = 'upper right', fontsize = 20)

plt.annotate('Min precipitation limit above which it is considered as "Moderate Shower"', fontstyle = 'italic', xy = (mdates.date2num(df.loc[15, 'timestamp']), 2), xytext = (mdates.date2num(df.loc[30, 'timestamp']), 4.5),
             arrowprops = dict(facecolor = 'black', shrink = 0.035, connectionstyle = 'arc3, rad = 0.3'), fontsize = 20)

plt.grid(alpha = 0.3, color = 'grey')

plt.show()
```

- From the above plot we can clearly see that there is exactly 0mm of "Total Precipitation" for most of the time. And cosidering the "Moderte Shower Limit", we can see only a few of them above the 2mm mark. This plot tells us that rainfall recorded in the month of August(2020) is less than the previous years. Although, August is considered to be the wettiest month of the year(in Delhi), the data deviates us from our dogmatic beliefs and we jump to the cocnlusion that this year(2020), Delhi has recorded less rainfall than the previous years.


- Moving on, we can see four attributes namely : "New Delhi Cloud Cover Total", "New Delhi Cloud Cover High [high cld lay]", "New Delhi Cloud Cover Medium [mid cld lay]", "New Delhi Cloud Cover Low [low cld lay]" i.e., "Total Cloud Cover", "High Cloud Cover", "Medium Cloud Cover" and "Low Cloud Cover" respectively. Before making any interpretations let us first know the exact meaning "Cloud Cover" and it's impact on climate.


- Cloud cover refers to the fraction of the sky obscured by clouds when observed from a particular location. Low cloud cover begins from the surface up to 2 km (at equator). On the other hand, medium cloud cover lies between 2-7 km (at equator). And lastly, high cloud cover is measured from 5-13 km (at equator).


- Cloud cover has a significant effect on the temperature forecast. The quantity of solar energy reaching the surface is a strong determinant of surface temperature. During daytime, clouds reduce the temperature, depending on the cloud thickness and cloud density but at night, clouds have the opposite effect on temperature.

```python
plt.figure(figsize=(50,10))

plt.title('Cloud Cover Analysis', fontsize=30)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Cloud Cover (%)', fontsize=15)

plt.plot(df['timestamp'], df['New Delhi Cloud Cover High [high cld lay]'],color = 'black')
plt.plot(df['timestamp'], df['New Delhi Cloud Cover Medium [mid cld lay]'],color = 'gray')
plt.plot(df['timestamp'], df['New Delhi Cloud Cover Low [low cld lay]'],color = 'silver')

plt.fill_between(df['timestamp'], df['New Delhi Cloud Cover High [high cld lay]'],color = 'black', alpha=0.4, label = 'High Cloud Cover')
plt.fill_between(df['timestamp'], df['New Delhi Cloud Cover Medium [mid cld lay]'],color = 'gray', alpha=0.4, label = 'Mid Cloud Cover')
plt.fill_between(df['timestamp'], df['New Delhi Cloud Cover Low [low cld lay]'],color = 'silver', alpha=0.5, label = 'Low Cloud Cover')

plt.ylim([-10, 110])

plt.text(mdates.date2num(df.loc[5, 'timestamp']), 48, '0 < Cloud Cover (%) < 100', fontstyle = 'italic', fontsize = 25, bbox = {'facecolor':'lightgray', 'pad':12, 'alpha':0.25})

plt.grid(alpha = 0.3, color = 'gray')

plt.legend(fontsize=20)

plt.show()
```

- From the above plot we can see that there is an uneven cloud cover for the entire week which tells us that the climate change in Delhi is not consistent in the month of August.


- Apart from analysing just the "Cloud Cover" we can also analyse the impact of "Cloud Cover" on other attributes. As we know that "Cloud Cover" is referred to the fraction of the sky obscured by clouds, so the attribute with the evident impact of "Cloud Cover" is "New Delhi Sunshine Duration" i.e., "Sunshine Duration".


- The relationship between cloud cover and sunshine duration is so obvious that it seems that there is no need to analyse it. As we know, during the day, the earth is heated by the sun. If skies are clear, more heat reaches the earth's surface. This leads to warmer temperatures. However, if skies are cloudy, some of the sun's rays are reflected off the cloud droplets back into space. Therefore, less of the sun's energy is able to reach the earth's surface, which causes the earth to heat up more slowly. This leads to cooler temperatures. During the night time, the cloud cover functions as a blanket and prevents heat from escaping from the Earth’s atmosphere.

Since one of the attributes is "Sunshine Duration" and it is easily understood that sunshine occurs during noon time so, we will only pick the data which consists of "New Delhi Cloud Cover Total" and "New Delhi Sunshine Duration" i.e., "Total Cloud Cover" and "Sunshine Duration" respectively but for those data values which are recorded at 12:00 hours.

- Apart from temperature, humidity also affects certain other attrbutes. And in order to have an incisive interpretation about the impact of humidity on other parameters, let us consider "New Delhi Diffuse Shortwave Radiation" i.e., "Diffuse Solar Radiation" and try to find out it's behaviour under the influence of "New Delhi Relative Humidity [2 m]" i.e., "Humidity".

```python
humidity_radiation_table = df.loc[df['hours']==12, ['timestamp', 'New Delhi Relative Humidity [2 m]', 'New Delhi Diffuse Shortwave Radiation']]
```

- Relative humidity is the ratio of the current absolute humidity to the highest possible absolute humidity. Absolute humidity is the mass of water vapor divided by the mass of dry air in a volume of air at a given temperature.
    

- Diffuse radiation describes the sunlight that has been scattered by molecules and particles in the atmosphere but that has still made it down to the surface of the earth.

```python
plt.figure(figsize = (30,10))

plt.title('Relative Humidity vs Diffuse Solar Radiation', fontsize=30)
plt.xlabel('Date', fontsize=15)

plt.plot(humidity_radiation_table['timestamp'], humidity_radiation_table['New Delhi Relative Humidity [2 m]'], color = 'black', alpha = 0.7, linewidth = 4, label = 'Relative Humidity')
plt.ylabel('Relative Humidity (%)', fontsize=15)

plt.legend(loc = 'upper left', fontsize = 20)


plt.twinx().plot(humidity_radiation_table['timestamp'], humidity_radiation_table['New Delhi Diffuse Shortwave Radiation'], color = 'yellow', alpha = 0.7, linewidth = 4, label = 'Diffuse Solar Radiation')
plt.ylabel('Diffuse Solar Radiation (W/m2)', fontsize=15)
plt.ylim([200, 310])

plt.legend(fontsize = 20, loc = (0.006,0.835))


plt.grid(alpha = 0.3, color = 'gray')

plt.show()
```

- From the above plot we can see that increase in "Relative Humidity" results in the increase in "Diffuse Solar Radiation". This is because more is the humidity, more will be the molecules and particles present in the atmosphere. This will further result in an increament in the "Diffuse Solar Radiation". This depicts an linear relationship between "Relative Humidity" and "Diffuse Solar Radiation" intensity.


- After learning about their behaviour I am curious about shortwave and longwave radiation. Firstly, we should know the exact difference between them. Shortwave radiation is nothing but visible light and on the other hand, longwave radiation is infrared light.


- Shortwave radiation contains a lot of energy than longwave radiation but has a shorter wavelength than longwave radiation. Solar energy enters our atmosphere as shortwave radiation in the form of ultraviolet rays and visible light. The sun emits shortwave radiation because it is extremely hot and has a lot of energy to give off.


- Shortwave radiation is further divided into two main categories namely direct shortwave radiation and diffuse shortwave radiation. Direct shortwave radiation is described as solar radiation traveling on a straight line from the sun down to the surface of the earth. And on the other hand, diffuse shortwave radiation is described as the sunlight that has been scattered by molecules and particles in the atmosphere but has still made it down to the surface of the earth.

```python

plt.figure(figsize = (30, 10))

plt.bar(df['timestamp'], df['New Delhi Shortwave Radiation'], width = 0.03, color = 'silver', alpha = 0.5, label = 'Total Shortwave Radiation')
plt.bar(df['timestamp'], df['New Delhi Direct Shortwave Radiation'], width = 0.03, color = 'cyan', alpha = 0.5, label = 'Direct Radiation')
plt.bar(df['timestamp'], df['New Delhi Diffuse Shortwave Radiation'], width = 0.03, color = 'greenyellow', alpha = 0.8, label = 'Diffuse Radiation')


plt.title('Shortwave Radiation', fontsize = 30)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Solar Radiation (W/m2)', fontsize = 16)
plt.ylim([0, 1000])
plt.grid(alpha = 0.3, color = 'gray')

plt.legend(fontsize = 20)




plt.tight_layout(pad = 4)

plt.show()
```

- The above plot tries to explain that the "Total Shortwave Radiation" is actually the sum of the other two attributes. But since it is a 2-Dimensional plot, we cannot clearly see the informative part. So, in order to understand it better we can plot the above graph in a 3-Dimensional representation.

```python
from mpl_toolkits.mplot3d import axes3d


fig = plt.figure(figsize = (50, 40))
ax = fig.add_subplot(111, projection = '3d')

x_val = mdates.date2num(df['timestamp'])

x_diffuse = x_val
y_diffuse = np.zeros(192)
z_diffuse = np.zeros(192)

dz_val_diffuse = df['New Delhi Diffuse Shortwave Radiation']

dx_diffuse = np.full((192), 0.01)
dy_diffuse = np.full((192), 0.065)
dz_diffuse = dz_val_diffuse

ax.bar3d(x_diffuse, y_diffuse, z_diffuse, dx_diffuse, dy_diffuse, dz_diffuse, color = 'greenyellow', alpha = 0.8)


x_direct = x_val
y_direct = np.full((192), 3)
z_direct = np.zeros(192)

dz_val_direct = df['New Delhi Direct Shortwave Radiation']

dx_direct = np.full((192), 0.01)
dy_direct = np.full((192), 0.065)
dz_direct = dz_val_direct

ax.bar3d(x_direct, y_direct, z_direct, dx_direct, dy_direct, dz_direct, color = 'cyan', alpha = 0.8)


x_total = x_val
y_total = np.full((192), 6)
z_total = np.zeros(192)

dz_val_total = df['New Delhi Shortwave Radiation']

dx_total = np.full((192), 0.01)
dy_total = np.full((192), 0.065)
dz_total = dz_val_direct

ax.bar3d(x_total, y_total, z_total, dx_total, dy_total, dz_total, color = 'silver', alpha = 0.8)



plt.title('Total Shortwave Radiation (Sum of Diffuse Radiation and Direct Radiation)', fontsize = 50, pad = 100)


plt.xlim(737639.9, 737648)
plt.ylim(0, 10)
ax.set_zlim(0, 1000)

plt.xlabel('Date', fontsize = 40, labelpad = 30)
ax.set_zlabel('Solar Radiation (W/m2)', fontsize = 40, labelpad = 30)



plt.tight_layout()


plt.show()
```

- From the above plot we can clearly see that "Total Shortwave Radiation" is exactly equal to the sum of "Direct Shortwave Radiation" and "Diffuse Shortwave Radiation".


- Solar radiation incoming from the sun affects the land too. As we know that soil is the main component of land, so let us now study about soil and it's constituents. In the data given, we can see two attributes related to soil namely : "New Delhi Soil Temperature [0-10 cm down]" and "New Delhi Soil Moisture [0-10 cm down]" i.e., "Soil Temperatutre" and "Soil Moisture".


- Soil moisture is referred as water content stored in the soil and in simple terms it is described as the water that is held in the spaces between soil particles.

```python
plt.figure(figsize = (30,10))

plt.title('Soil Moisture vs Soil Temperature', fontsize = 30)
plt.xlabel('Date', fontsize = 16)

plt.plot(df['timestamp'], df['New Delhi Soil Moisture [0-10 cm down]'], color = 'gold', linewidth = 4, label = 'Soil Moisture')
plt.ylabel('Soil Moisture (fraction)', fontsize = 16)

plt.legend(fontsize = 20, loc = 'upper center')

plt.twinx().plot(df['timestamp'], df['New Delhi Soil Temperature [0-10 cm down]'], color = 'magenta', linewidth = 4, label = 'Soil Temperature')
plt.ylabel('Soil Temperature (°C)', fontsize = 16)

plt.legend(fontsize = 20, loc = (0.44, 0.83))

plt.grid(alpha = 0.3, color = 'grey')

plt.show()
```

- Similar to the EDA performed above we can clearly see that increase in moisture content of soil decreases the soil temperature. And therefore "Soil Moisture" and "Soil Temperature" are inversely related to each other.


- We have discussed enough about temperature, humidity, moisture, etc. Now let us discuss about some of the factors affecting rainfall and lightening. One of the most significant factors affecting thunderstorm is "New Delhi CAPE [180-0 mb above gnd]" i.e., "Convective Available Potential Energy" commonly abbreviated as CAPE.

```python
plt.figure(figsize = (30,10))

plt.title('CAPE Analysis', fontsize = 30)
plt.xlabel('Date', fontsize = 16)

plt.plot(df['timestamp'], df['New Delhi CAPE [180-0 mb above gnd]'], label = 'CAPE (Convective Available Potential Energy)', color = 'orange', linewidth = 4)
plt.ylabel('CAPE (J/kg)', fontsize = 16)

plt.ylim([0, 3000])

plt.fill_between(df['timestamp'], 1000, df['New Delhi CAPE [180-0 mb above gnd]'], where = df['New Delhi CAPE [180-0 mb above gnd]'] > 1000, color = 'bisque', alpha = 0.5)
plt.grid(alpha = 0.3, color = 'grey')

plt.axhline(y = 1000, color = 'red', linestyle = ':', linewidth = 2)
plt.axhspan(1000, 2500, color = 'bisque', alpha = 0.25, label = 'Area of Moderate Instability')

plt.legend(fontsize = 20, loc = 'upper left')



plt.annotate('Min CAPE value above which it is considered as "Moderate Instability"', fontstyle = 'italic', xy = (mdates.date2num(df.loc[17, 'timestamp']), 1000), xytext = (mdates.date2num(df.loc[30, 'timestamp']), 1900),
             arrowprops = dict(facecolor = 'black', shrink = 0.01, connectionstyle = 'arc3, rad = 0.3'), fontsize = 20)


plt.show()
```

- The above plot tells us that the month of August has been a good month in terms of moderate thunder and lightning. The above plot justifies the moderate weather conditions in the month of August as there is an average CAPE value at 958 J/kg whereas the minimum value being 185 J/kg and the maximum being 1706 J/kg. Since all the three values are below 2500 J/kg. Therefore it ensures a moderate instability which is quite good in the month of August.
